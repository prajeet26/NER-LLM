```python
def get_completion(query: str, model, tokenizer) -> str:
  device = "cuda:0"

  prompt_template = """
  <start_of_turn>user
  Translate German to English
  {query}
  <end_of_turn>\n<start_of_turn>model


  """
  prompt = prompt_template.format(query=query)

  encodeds = tokenizer(prompt, return_tensors="pt", add_special_tokens=True)

  model_inputs = encodeds.to(device)


  generated_ids = model.generate(**model_inputs, max_new_tokens=1000, do_sample=True, pad_token_id=tokenizer.eos_token_id)
  # decoded = tokenizer.batch_decode(generated_ids)
  decoded = tokenizer.decode(generated_ids[0], skip_special_tokens=True)

  return (decoded)
```

```python
import csv
import json

def csv_to_json(csv_file_path, output_file_path):
    data = []

    with open(csv_file_path, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            user_input = {"role": "user", "content": row["combined"]}
            assistant_output = {"role": "assistant", "content": row["entities"]}
            data.append(user_input)
            data.append(assistant_output)

    with open(output_file_path, 'w') as jsonfile:
        json.dump(data, jsonfile, indent=4)

# Replace 'input.csv' with the path to your CSV file
# Replace 'output.json' with the desired output JSON file path
csv_to_json('/content/combtext_final.csv', '/content/combtext.json')
```

```python
import json
import random
from datasets import Dataset

def split_json_data(input_file_path, train_ratio=0.8):
    with open(input_file_path, 'r') as jsonfile:
        data = json.load(jsonfile)

    # Shuffle the data
    random.shuffle(data)

    # Split the data into train and test sets
    train_test_split_index = int(len(data) * train_ratio)
    train_data = data[:train_test_split_index]
    test_data = data[train_test_split_index:]

    return train_data, test_data

# Replace 'output.json' with the path to your JSON file
train_data, test_data = split_json_data('/content/combtext.json')

# Replace 'save_path' with the desired path to save the train and test datasets
# Replace 'file_name_train_chatml' and 'file_name_val_chatml' with the desired file names for the train and test datasets
save_path = "/content/"
file_name_train_chatml = "train.json"
file_name_val_chatml = "test.json"

with open(save_path + file_name_train_chatml, 'w') as f:
    json.dump(train_data, f, indent=4)

with open(save_path + file_name_val_chatml, 'w') as f:
    json.dump(test_data, f, indent=4)

with open('/content/train.json', 'r') as f:
    train_data = json.load(f)

with open('/content/test.json', 'r') as f:
    test_data = json.load(f)

dataset_train = Dataset.from_list(train_data)
dataset_test = Dataset.from_list(test_data)
```

```python
max_length = 2300
tokenizer.pad_token = tokenizer.eos_token
def generate_and_tokenize_prompt(text):
    result = tokenizer(
        text,
        truncation=True,
        max_length=max_length,
        padding="max_length",
    )
    result["labels"] = result["input_ids"].copy()
    return result

tokenized_train_dataset = dataset_train.map(
    lambda example: generate_and_tokenize_prompt(example["content"]),
    batched=True,
)

tokenized_test_dataset = dataset_test.map(
    lambda example: generate_and_tokenize_prompt(example["content"]),
    batched=True,
)
```

```python
import transformers
from datetime import datetime

project = "LLM-NER"
base_model_name = "phi"
run_name = base_model_name + "-" + project
output_dir = "./" + run_name

training_arguments = transformers.TrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=1,
    learning_rate=2e-4,
    lr_scheduler_type="cosine",
    save_strategy="steps",
    gradient_checkpointing=True,
    logging_steps=30,
    max_steps=90,
    fp16=True,
    num_train_epochs=1,
    report_to="wandb",
    evaluation_strategy="steps",
    logging_dir="./logs",
)

trainer = transformers.Trainer(
    model=model,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_test_dataset,
    args=training_arguments,
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False),
)

model.config.use_cache = False
trainer.train()
```
